PART B TRAINING MLP CLASSIFIER

```python
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import log_loss
from sklearn.model_selection import train_test_split

# Spliting training data into a smaller train values
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# mlp parameters
layer_sizes = [128, 64, 32]
batch_size = 128
learning_rate = 2e-5
max_iter = 100

# Activation functions to see
given = ['logistic', 'tanh', 'relu', 'identity']

# Dictionary to store loss histories for each activation function
train_dict = {}
val_dict = {}

# Train a model for each activation function
for i in given:
    mlp = MLPClassifier(hidden_layer_sizes=layer_sizes,
                        max_iter=1,  # to see loss per epoch
                        solver='adam',
                        batch_size=batch_size,
                        learning_rate_init=learning_rate,
                        activation=i,
                        warm_start=True)  # Allows incremental training across iterations

      # Track losses for plotting
    train_losses = []
    val_losses = []

    # Train for 100 epochs
    for epoch in range(max_iter):
        mlp.fit(X_train_sub, y_train_sub)  # Train the model for 1 epoch
        train_loss = log_loss(y_train_sub, mlp.predict_proba(X_train_sub))  # Calculate training loss
        val_loss = log_loss(y_val, mlp.predict_proba(X_val))  # Calculate validation loss

        train_losses.append(train_loss)
        val_losses.append(val_loss)

        print(f"Epoch {epoch+1}/{max_iter} - Activation: {i} - Train Loss: {train_loss:.5f} - Val Loss: {val_loss:.5f}") # upto 5 decimal places

    # Store the losses for plotting later
    train_dict[i] = train_losses
    val_dict[i] = val_losses

# Plot training and validation loss curves for each activation function
plt.figure(figsize=(15, 10))
for j in given:
    plt.plot(train_dict[j], label=f'Train Loss ({j})')
    plt.plot(val_dict[j], label=f'Val Loss ({j})', linestyle='--')

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and Validation Loss for Different Activation Functions")
plt.legend()
plt.show()

# Evaluate each model on the test set and determine the best-performing activation function
test_accuracies = {}
for k in given:
    mlp = MLPClassifier(hidden_layer_sizes=layer_sizes,
                        max_iter=max_iter,
                        solver='adam',
                        batch_size=batch_size,
                        learning_rate_init=learning_rate,
                        activation=k)
    mlp.fit(X_train, y_train)
    test_accuracy = mlp.score(X_test, y_test)
    test_accuracies[k] = test_accuracy
    print(f"Test accuracy for activation {k}: {test_accuracy:.4f}")

# Find and display the best activation function
best_activation = max(test_accuracies, key=test_accuracies.get)
print(f"Best performing activation function on the test set: {best_activation} with accuracy {test_accuracies[best_activation]:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier

# Best activation function from part 2
best_activation = 'relu'

# Define the hyperparameter grid
param_grid = {
    'solver': ['adam', 'sgd', 'lbfgs'],  # Solvers to try
    'learning_rate_init': [1e-4, 1e-5, 1e-6],  # Different learning rates
    'batch_size': [64, 128, 256],  # Different batch sizes
    'hidden_layer_sizes': [(128, 64, 32)],  # The layer sizes we use
    'max_iter': [100]
}

mlp = MLPClassifier(activation=best_activation, random_state=42)
grid_search = GridSearchCV(estimator=mlp, param_grid=param_grid, cv=3, n_jobs=-1, verbose=1)

# Fitting grid search
grid_search.fit(X_train, y_train)

print("Best Hyperparameters from Grid Search:")
print(grid_search.best_params_)

# Evaluate the best model on the test set
best_mlp = grid_search.best_estimator_
test_accuracy = best_mlp.score(X_test, y_test)
print(f"Test Accuracy of Best Model: {test_accuracy:.4f}")
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# for our nn
layer_sizes = [128, 64, 32, 64, 128]
learning_rate = 2e-5
max_iter = 100
batch_size = 128

# initialise 2 given relu and identity
mlp_relu = MLPRegressor(hidden_layer_sizes=layer_sizes, activation='relu', solver='adam',
                        learning_rate_init=learning_rate, max_iter=1, batch_size=batch_size, random_state=42, warm_start=True)

mlp_identity = MLPRegressor(hidden_layer_sizes=layer_sizes, activation='identity', solver='adam',
                            learning_rate_init=learning_rate, max_iter=1, batch_size=batch_size, random_state=42, warm_start=True)

# To store the loss curves
train_losses_relu, val_losses_relu = [], []
train_losses_identity, val_losses_identity = [], []

# loop for training
for model, activation, train_losses, val_losses in [
    (mlp_relu, 'relu', train_losses_relu, val_losses_relu),
    (mlp_identity, 'identity', train_losses_identity, val_losses_identity)
]:
    for epoch in range(max_iter):
        # Train model for one epoch
        model.fit(X_train, X_train)
        train_loss = mean_squared_error(X_train, model.predict(X_train))
        train_losses.append(train_loss)
        val_loss = mean_squared_error(X_test, model.predict(X_test))
        val_losses.append(val_loss)

# Plotting the training and validation losses
plt.figure(figsize=(12, 6))
plt.plot(train_losses_relu, label='Training Loss (ReLU)', color='blue')
plt.plot(val_losses_relu, label='Validation Loss (ReLU)', color='cyan')
plt.plot(train_losses_identity, label='Training Loss (Identity)', color='red')
plt.plot(val_losses_identity, label='Validation Loss (Identity)', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss per Epoch for ReLU and Identity Activations')
plt.show()

# Visualize the generated outputs for 10 test samples
fig, axes = plt.subplots(2, 10, figsize=(20, 5))
test_samples = X_test[:10]

for i, (model, activation) in enumerate([(mlp_relu, 'ReLU'), (mlp_identity, 'Identity')]):
    generated_images = model.predict(test_samples)
    for j in range(10):
        axes[i, j].imshow(generated_images[j].reshape(28, 28), cmap='gray')
        axes[i, j].axis('off')
        if j == 0:
            axes[i, j].set_ylabel(activation, size=14, labelpad=20)

plt.suptitle('Generated Images Using ReLU and Identity Activation Functions')
plt.tight_layout()
plt.show()
```
